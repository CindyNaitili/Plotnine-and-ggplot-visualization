```python
#Multiline plot

work_years=[1,2,3,4]

ml_engineer_income=[10000,12000,14000,16000]
sde_income=[8000,11000,10000,17000]

income_df=pd.DataFrame({"Years at work": work_years, "ML engineer income":ml_engineer_income,"SDE Income":sde_income})
```

```python
income_df.head()
```

```python
(ggplot()+
 plotnine.geom_line(data=income_df,mapping=aes(x='Years at work', y='ML engineer income', color="'blue'"))+
 plotnine.geom_line(data=income_df,mapping=aes(x='Years at work', y='SDE Income', color="'red'"))+
 plotnine.labs(title="Multiline plot from ML and SDE Engineer Income/years")+
 plotnine.scale_color_identity(guide='legend', name="Legend", breaks=['blue','red'],labels=['ML Engineer Income','SDE Income'])).draw()
```

```python
#bar plot

df = sns.load_dataset('titanic')
df.head()
```

```python
(plotnine.ggplot(data=df.dropna(), mapping= aes(x='pclass', y='fare', fill='sex'))+
 plotnine.geom_col()+
 plotnine.labs(title="Bar Plot of Passenger class vs fare")).draw()
```

```python
#Scatter Plot

scatter_plot=ggplot(data=df, mapping=aes(x='pclass', y='fare'))

(scatter_plot +
 plotnine.labs(title="Scatter Plot for Passenger Class vs Fare")+
 plotnine.geom_point(shape='o',size=7,color="green",alpha=0.7)).draw()
```

```python
#Histogram

hist = ggplot(data=df, mapping=aes(x='alive'))

(hist +
 plotnine.labs(title="Histogram of Passenger status")+
 plotnine.geom_histogram()+
 plotnine.geom_rug()).draw()

```

```python
#Density Plot

kde = ggplot(data=df,mapping = aes("age"))

(kde+
 plotnine.geom_density()+
 plotnine.labs(title="Density Plot for Age of Passenger")).draw()
```

```python
#Box Plot

bp = ggplot(data=df, mapping=aes(x="embarked", y="fare"))

(bp+
 plotnine.geom_boxplot()+
 plotnine.labs(title="Box Plot for Embarked vs Fare")).draw()
```

```python
#Violin Plot

vp = ggplot(data=df, mapping=aes(x='sex', y='age'))

(vp+
 plotnine.geom_violin()+
 plotnine.labs(title="Violin Plot of Passenger Age by Gender")+
 plotnine.geom_jitter(position=plotnine.position_jitter(0.3))).draw()
```
